```python
def db_connection(funct):
    def inner():
        print("connection establishment")
        funct()
        print("connection termination")
    return inner
        
```

```python
@db_connection
def copper_data_store():   
    print("data transfer for copper")
@db_connection 
def silver_data_store():
    print("data transfer for silver")
@db_connection
def gold_data_store():   
    print("data transfer for gold")
   
```

```python
copper_data_store()
silver_data_store()
gold_data_store()
```
