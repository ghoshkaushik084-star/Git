```python
User ID Password example
```

```python
email = input ("Enter your email:")
password = input ("Enter your password:")
if email == "kaushik@gmail.com" and password == "1234":
    print ("Welcome")
elif email == "kaushik@gmail.com" and password != "1234":
    password = input ("Re-enter your password")
    if password == "1234":
        print ("Finally correct")
    else:
        print ("Still incorrect")
else:
    print ("Incorrect credentials")
```

```python
3. Take marks of three subjects;
If average ≥ 50 → “Pass”
If average ≥ 80 → “Distinction”
Else “Fail”.
```

```python
mark1 = float(input("Enter number for subject1"))
mark2 = float(input("Enter number for subject2"))
mark3 = float(input("Enter number for subject3"))
average = (mark1+mark2+mark3)/3
if average >=50:
    if average>=80:
       print("Distinction")
    else:
       print("Pass")
else:
    print("Fail")
```

```python
4. Input a year and check:
      If divisible by 4
         If divisible by 100
              If divisible by 400 → “Leap Year”
              Else “Not Leap Year”
         Else “Leap Year”
      Else “Not Leap Year”
```

```python
year = int(input("Enter a year"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print("Leap year")
        else:
            print("Not Leap year")
    else:
        print("Leap year")
else:
    print("Not Leap year")
```

```python
5. Take an integer and check if it’s positive and multiple of 5, else print the respective reason (e.g., “Not multiple of 5” or “Negative number”).
```

```python
num = int(input("Enter an integer"))
if num>0:
    if num%5==0:
        print("The number is positive and multiple of 5")
    else:
        print("The number is positive but not multiple of 5")
else:
    print("Negative number")
```

```python
1. Email fetching valid, invalid
```

```python
emails = ['test@gmail.com', 'hdfc@yahoo.com', 'sbi@redif.com', 'panch@bab.com', 'mumbai@gmail.com', 'bangalore@hot.com', 'pune@gmail.com']
```

```python
emails
```

```python
valid_emails = []
invalid_emails = []
for email in emails:
    if email.endswith('@gmail.com'):
        valid_emails.append(email)
    else:
        invalid_emails.append(email)
print('valid emails:', valid_emails)
print('invalid emails:', invalid_emails)        
```
